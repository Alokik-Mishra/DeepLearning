# ECBM E4040 - Assignment 2 - Task 4: Data Augmentation

One important factor in neural network training is the size of the traininig set. Since it is often not possible to get a clean and large enough dataset for training, one way to improve the network's robustness and generalization ability is to create 'fake' data by injecting random noise or doing random transformations on the available data. A technique which implements this strategy is called __data augmentation__, and has shown to be very effective.

One thing to remember, when you augment your data, is to never change the correct label of a sample. For example, for hand-written digit dataset, flipping a letter 'b' ends up looking like a letter 'd', but you must keep the label for 'b'. So please choose the best augmentation methods for your dataset.

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

# Import modules
from __future__ import print_function
import numpy as np
from ecbm4040.cifar_utils import load_data
import matplotlib.pyplot as plt
import tensorflow as tf
```

## Load Data

```python
X_train, y_train, X_test, y_test = load_data()

num_train = 49000
num_val = 1000
num_test = 10000
num_dev = 128

# The development set is used for augmentation practices.
mask = np.random.choice(num_train, num_dev, replace=False)
X_dev = X_train[mask]
y_dev = y_train[mask]

# Seperate Training set into a training set and a validation set
X_val = X_train[num_train:]
y_val = y_train[num_train:]
X_train = X_train[:num_train]
y_train = y_train[:num_train]

print(X_train.shape, X_val.shape, X_test.shape, X_dev.shape)
```

## Part 1: Visualization

### Visualize some original images


```python

X_dev_reshaped = X_dev.reshape([128,3,32,32]).transpose((0,2,3,1))
X_dev_reshaped.shape
mask = np.random.choice(X_dev.shape[0], 16, replace=False)
X_vis = X_dev_reshaped[mask, :, :, :]

np.random.seed(123)
r = 4
f, axarr = plt.subplots(r, r, figsize=(8,8))
for i in range(r):
    for j in range(r):
        img = X_vis[r*i+j]
        axarr[i][j].imshow(img, cmap="gray")
X_dev_reshaped.shape
```

## Part 2: Automatic batch generator

```python
from ecbm4040.image_generator import ImageGenerator
```

```python

Dev = ImageGenerator(x = X_dev_reshaped, y = y_dev)
Count = Dev.next_batch_gen(batch_size = 16)
Dev.show(next(Count)[0])
```

### Translation


```python
Dev = ImageGenerator(x = X_dev_reshaped, y = y_dev)
Dev_t = Dev.translate(5, 5)
Dev_t = ImageGenerator(Dev_t[0], Dev_t[1])
Count = Dev_t.next_batch_gen(batch_size = 16)
Dev_t.show(next(Count)[0])
```

### Rotation


```python
Dev = ImageGenerator(x = X_dev_reshaped, y = y_dev)
Dev_r = Dev.rotate(angle = 45)
Dev_r = ImageGenerator(Dev_r[0], Dev_r[1])
Count = Dev_r.next_batch_gen(batch_size = 16)
Dev_r.show(next(Count)[0])
```

### Flipping (horizontal and vertical)


```python
Dev = ImageGenerator(x = X_dev_reshaped, y = y_dev)
Dev_f = Dev.flip(mode = 'hv')
Dev_f = ImageGenerator(Dev_f[0], Dev_f[1])
Count = Dev_f.next_batch_gen(batch_size = 16)
Dev_r.show(next(Count)[0])
```

### Add Noise



```python
Dev = ImageGenerator(x = X_dev_reshaped, y = y_dev)
Dev_n = Dev.add_noise(portion = 0.5, amplitude = 40)
Dev_n = ImageGenerator(Dev_n[0], Dev_n[1])
Count = Dev_n.next_batch_gen(batch_size = 16)
Dev_n.show(next(Count)[0])
```

## Part 3: Data Augmentation + LeNet


```python
from ecbm4040.neuralnets.cnn import my_training_task4
tf.reset_default_graph()

num_training = 49000
num_validation = 1000

X_val = X_train[-num_validation:, :]
y_val = y_train[-num_validation:]

X_train = X_train[:num_training, :]
y_train = y_train[:num_training]

# Preprocessing: subtract the mean value across every dimension for training data, and reshape it to be RGB size
mean_image = np.mean(X_train, axis=0)
X_train = X_train.astype(np.float32) - mean_image.astype(np.float32)
X_val = X_val.astype(np.float32) - mean_image

X_train = X_train.reshape([-1,32,32,3]) / 255
X_val = X_val.reshape([-1,32,32,3]) / 255

my_training_task4(X_train, y_train, X_val, y_val, 
         conv_featmap=[6,18],
         fc_units=[128],
         conv_kernel_size=[3,5],
         pooling_size=[2,2],
         l2_norm=0.05,
         seed=235,
         learning_rate = 1e-2,
         epoch=25,
         batch_size=735,
         verbose=True,
         pre_trained_model=None)
```

We see a drop in the performance of the network with the two additions to the data, one being a rotation and the other being a horizontal and verticle flip. The performance of the model drop almost by half from 65 to 33% validation accuracy.

We see that the difference in loss is not so high while the difference in the validation accuracy is high. This likely indicates overfitting. The images are pretty low definition and not extremely different (particualrily some animals), and so training on mutlitple images which are rotated and flipped might lead to overfitting.
